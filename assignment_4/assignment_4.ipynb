```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup

header = [ 'title', 'url', 'rate']
data = []
# Display and store away 2 pages of scrapped data from startupbeat.hkej.com
headers = {"User-Agent":'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.71 Safari/537.36'}
for i in range(0, 101, 25):
    quote_page = requests.get(
        f'https://www.douban.com/doulist/2422549/?start={i}&sort=seq&playable=0&sub_type=', headers=headers)
    print("\n***** Page " + str(i) + " in action *****")
    soup = BeautifulSoup(quote_page.text, 'html.parser')
    target = soup.find_all("div", attrs={"class": "doulist-item"})
    print(target)
    for game in target:

        title = game.find("div", attrs={"class": "title"}).a.text.strip()
        url = game.find("div", attrs={"class": "post"}).a.get('href')
#         details = game.find("div", attrs={"class": "abstract"}).get_text().strip()
        rate = game.find("span", attrs={"class": "rating_nums"}).get_text()
        print(title)
        print(url)
#         print(details)
        print(rate)
        data.append((title, url, rate))

df = pd.DataFrame(data,
                  columns=header
                  )
df.to_csv("game.csv", encoding='utf-8-sig')
df
```
